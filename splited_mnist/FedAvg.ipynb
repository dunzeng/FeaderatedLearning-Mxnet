## FedAvg实验数据处理
将60,000张mnist图片分为600张/client，即100个client
保证数据为IID

mnist数据分布：
[5923, 6742, 5958, 6131, 5842, 5421, 5918, 6265, 5851, 5949]

```python
import mxnet as mx
import numpy as np
```

```python
mnist = mx.test_utils.get_mnist()
```

```python
data = mnist['train_data']
label = mnist['train_label']
```

```python
data.shape
```

```python
data[0:60].shape
```

```python
label.shape
```

```python
client_ = {}
for i in range(100):
    client_["data" + str(i)] = data[600*i: 600*i+600]
    client_["label" + str(i)] = label[600*i: 600*i+600]
for i in range(100):
    np.save(".\\FedAvg_data\\train_data"+str(i)+".npy",client_["data"+str(i)])
    np.save(".\\FedAvg_data\\train_label"+str(i)+".npy",client_["label"+str(i)])
```

```python
mnist_data = []
for i in range(10):
    mnist_data.append(np.load("data"+str(i)+".npy"))
```

```python
for idx in range(10):   #遍历label
    for i in range(100): #遍历client
        data = mnist_data[idx][i*60:(i+1)*60]
        data = data.tolist()
        client_["data"+str(i)] += data
        client_["label"+str(i)] += len(data)*[idx]
```
